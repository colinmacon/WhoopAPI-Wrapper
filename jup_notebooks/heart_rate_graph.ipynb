```python
import pandas as pd 
from IPython.display import display, HTML

import plotly.graph_objs as go
```

```python
# Reading in our df 
df = pd.read_pickle("../data/clean/hr_data/2021-12-27_2022-08-01.csv")
```

```python
# View the entire df head for ref 
display(HTML(df.head().to_html()))
```

#### Filter Dates as graphing all the Heart Rate Series Data could freeze your machine
> hr_series() in VisbaleWhoop

```python
# Filter dates 

# All data after select date
#df = df.loc[df['date'] > '2022-05-06'] # cjx data 

# Between two dates 
df = df.loc[df["time"].between("2022-07-25 01:00:00.000", "2022-07-25 23:59:00.000")]
```

## HR Series Plot

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
x=df.time,
y=df["bpm"],
mode="markers+lines",
name = "Heart Rate",
text = "HR:" + df['time'].astype('string') + " - " + df['bpm'].astype('string'),
hoverinfo = 'text'))

fig.update_xaxes(title='Date')
fig.update_yaxes(title='Heart Rate')
fig.update_layout(title_text='Heart Rate Series Data', title_x=0.5)


fig.show()
```
